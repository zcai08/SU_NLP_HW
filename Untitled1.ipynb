```python
import nltk
import re
from nltk import FreqDist
from nltk.collocations import *
from nltk.stem import WordNetLemmatizer
nltk.corpus.gutenberg.fileids()
print(nltk.corpus.gutenberg.fileids())
```

```python
file0 =nltk.corpus.gutenberg.fileids()[0]
file1 = nltk.corpus.gutenberg.fileids()[4]
print(file0,file1)
```

```python
def alpha_filter(w):
  # pattern to match a word of non-alphabetical characters
    pattern = re.compile('^[^a-z]+$')
    if (pattern.match(w)):
        return True
    else:
        return False
```

```python
text0 = nltk.corpus.gutenberg.raw(file0)
tokens0 = nltk.word_tokenize(text0) 
words0 = [w.lower( ) for w in tokens0]
text1 = nltk.corpus.gutenberg.raw(file1)
tokens1 = nltk.word_tokenize(text1) 
words1 = [w.lower( ) for w in tokens1]
print(len(words0),len(words1))
```

```python
lemmatizer = WordNetLemmatizer()
lemmatizedwords0 = [lemmatizer.lemmatize(w) for w in words0]
lemmatizedwords1 = [lemmatizer.lemmatize(w) for w in words1]
```

```python
alphawords0 = [w for w in lemmatizedwords0 if not alpha_filter(w)]
alphawords1 = [w for w in lemmatizedwords1 if not alpha_filter(w)]
```

```python
nltkstopwords = nltk.corpus.stopwords.words('english')
morestopwords = ['could','would','might','must','need','sha','wo','y',"'s","'d","'ll","'t","'m","'re","'ve","n't"]
stopwords = nltkstopwords + morestopwords
stoppedwords0 = [w for w in alphawords0 if not w in stopwords]
stoppedwords1 = [w for w in alphawords1 if not w in stopwords]
```

```python
ndist0 = FreqDist(stoppedwords0)
nitems0 = ndist0.most_common(50)
for item in nitems0:
    print (item[0], '\t', item[1]/len(words0))
```

```python
ndist1 = FreqDist(stoppedwords1)
nitems1 = ndist1.most_common(50)
for item in nitems1:
    print (item[0], '\t', item[1]/len(words1))
```

```python
bigram_measures = nltk.collocations.BigramAssocMeasures()
finder0 = BigramCollocationFinder.from_words(words0)
finder0.apply_word_filter(alpha_filter)
finder0.apply_word_filter(lambda w: w in stopwords)
scored = finder0.score_ngrams(bigram_measures.raw_freq)
for bscore in scored[:50]:
    print (bscore)
```

```python
finder0.apply_freq_filter(5)
scored = finder0.score_ngrams(bigram_measures.pmi)
for bscore in scored[:50]:
    print (bscore)
```

```python
finder1 = BigramCollocationFinder.from_words(words1)
finder1.apply_word_filter(alpha_filter)
finder1.apply_word_filter(lambda w: w in stopwords)
scored = finder1.score_ngrams(bigram_measures.raw_freq)
for bscore in scored[:50]:
    print (bscore)
```

```python
finder1.apply_freq_filter(3)
scored = finder1.score_ngrams(bigram_measures.pmi)
for bscore in scored[:50]:
    print (bscore)
```

```python
trigram_measures = nltk.collocations.TrigramAssocMeasures()
finder2 = TrigramCollocationFinder.from_words(lemmatizedwords0)
finder2.apply_word_filter(alpha_filter)
finder2.apply_word_filter(lambda w: w in stopwords)
scored = finder2.score_ngrams(trigram_measures.raw_freq)
for bscore in scored[:50]:
    print (bscore)
```

```python
scored = finder2.score_ngrams(trigram_measures.pmi)
for bscore in scored[:50]:
    print (bscore)
```
